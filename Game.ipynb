```python
import random as r
```

```python


# Game set up
print("Welcome to the Dungeon Escape Game!")
print("You are an adventurer trapped in a dungeon.")
print("You must find the key to unlock the door and escape.")
print("Beware, there are monsters lurking in the shadows.")

# Initialize variables
has_key = False
monster_encountered = False

while True:
    # Give player options
    print("What would you like to do? (Enter the number)")
    print("1. Search for the key.")
    print("2. Check for monsters.")
    print("3. Attempt to open the door.")
    choice = input()

    # Handle player choice
    if choice == "1":
        print("You search the dungeon for the key.")
        if r.randint(0, 1) == 0:
            print("You found the key!")
            has_key = True
        else:
            print("The key is not here.")
    elif choice == "2":
        print("You listen carefully for any signs of monsters.")
        if r.randint(0, 1) == 0:
            print("You hear a growl nearby.")
            monster_encountered = True
        else:
            print("You hear nothing.")
    elif choice == "3":
        if has_key:
            print("You use the key to unlock the door and escape the dungeon!")
            break
        else:
            print("The door is locked. You need to find the key.")
    else:
        print("Invalid choice. Please enter a number 1-3.")

    # Check for monster encounters
    if monster_encountered:
        print("A monster appears!")
        if r.randint(0, 1) == 0:
            print("You defeat the monster and continue your search.")
            monster_encountered = False
        else:
            print("The monster kills you. Game over.")
            break
```
