```python
def welcome():
    print("Thank you for using our rent qualification calculator")
    print("Times are hard, and we understand that you don't want to waste money on application and admin fees for a denied application.")
    name = input("That's why we're here! Tell us your name and we'll get started: ")
    print("Thank you, " + name + ". We'll need to ask you a few questions next.")
    return name

def proposed_rent(name):
    print("First, we'll need the rent for the apartment you're applying for, " + name + ".")
    Rent = int(input("Enter your rent: "))
    rent_with_utilities = Rent + 200
    triple_rent = Rent * 3
    return triple_rent, rent_with_utilities, Rent

def monthly_income(triple_rent):
    print("Next, we'll need your GROSS monthly income.")
    income = int(input("Enter your gross monthly income: "))
    triple_income = income * 3
    cosigner = ""
    if triple_income < triple_rent:
        cosigner = input("Will you be able to get a cosigner if needed? (Yes or No only): ")
    return triple_income, cosigner

def Credit(Rent):
    print("We'll need your credit next.")
    credit = int(input("Enter your credit: "))
    deposit = ""
    if 500 <= credit < 630:
        deposit = input("Will you be able to put down a $" + str(Rent) + " deposit? (Yes or No only): ")
    return credit, deposit

def Rental_History():
    print("Now we'll ask about your rental/criminal history")
    rental_criminal_history = input("Do you have any evictions, bankruptcies, or felonies? (Yes or No only): ")
    return rental_criminal_history

def Decision_tree(triple_rent, triple_income, cosigner, credit, deposit, rental_criminal_history):
    Approval = "Congratulations, You've been approved for your new home!"
    Denial = "Unfortunately, you were not approved for your desired rent."
    suggested_income = f"Unfortunately, you were not approved for your desired rent, the rent you would be eligible for is ${triple_income / 3}."


    #if there's any evictions or such deny automatically
    # do not continue
    
    if rental_criminal_history == 'Yes':
        print(Denial)
    else:
        # if triple_income is less than rent
        # and they have have a cosigner, high credit
        # or deposit, approve
        if triple_income < triple_rent:
            if cosigner == 'Yes' or credit > 670 or deposit == 'Yes':
                print(Approval)
            # deny and print reason for
            # denial is low income
            else:
                print(suggested_income)
        # if triple_income is greater than rent
        # and credit is good approve, else deny
        elif triple_income > triple_rent:
            if credit > 670:
                print(Approval)
            else:
                print(Denial)

def main():
    name = welcome()
    triple_rent, rent_with_utilities, Rent = proposed_rent(name)
    triple_income, cosigner = monthly_income(triple_rent)
    credit, deposit = Credit(Rent)
    rental_criminal_history = Rental_History()
    Decision_tree(triple_rent, triple_income, cosigner, credit, deposit, rental_criminal_history)

if __name__ == "__main__":
    main()
```
